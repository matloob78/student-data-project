```python
# student info
student_id = 1
student_name = "Ali"

# subjects list (3â€“4 subjects)
subjects = [
    {"subject": "Geography", "marks": 80},
    {"subject": "English", "marks": 75},
    {"subject": "Science", "marks": 90},
    {"subject": "Urdu", "marks": 70}
]

# loop and add rows
for s in subjects:
    result = "Pass" if s["marks"] >= 50 else "Fail"
    
    new_row = {
        "id": student_id,
        "name": student_name,
        "subject": s["subject"],
        "marks": s["marks"],
        "result": result
    }
    
    df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

# save
df.to_csv("students_clean.csv", index=False)

print(df.tail(10))
```
